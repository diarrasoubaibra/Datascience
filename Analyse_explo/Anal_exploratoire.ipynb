**visualisation de la distribution des variables qualitatives et quantitatives en fonction de la variable 'CŒUR'**

```python
sns.pairplot(df[['AGE', 'FCMAX', 'CŒUR']], hue='CŒUR')
plt.title('Pair Plot pour AGE')
plt.show()
```

```python
df.head()
```

```python
# Liste des colonnes sauf 'CŒUR'
variables_dintérêt = [col for col in df.columns if col != 'CŒUR']

# Analyse bivariée pour chaque variable
for variable in variables_dintérêt:
    if df[variable].dtype == 'object':
        sns.countplot(x=variable, hue='CŒUR', data=df)
        plt.title(f'Distribution de CŒUR en fonction de {variable}')
        plt.show()
    else:
        sns.boxplot(x='CŒUR', y=variable, data=df)
        plt.title(f'Relation entre CŒUR et {variable}')
        plt.show()
```

```python
# Analyse bivariée pour chaque variable d'intérêt
for variable in variables_dintérêt:
    if df[variable].dtype == 'object':  # Variable qualitative
        sns.countplot(x=variable, hue='CŒUR', data=df)
        plt.title(f'Distribution de CŒUR en fonction de {variable}')
        plt.show()
    else:  # Variable quantitative
        # Utiliser un histogramme
        plt.hist([df[df['CŒUR'] == 0][variable], df[df['CŒUR'] == 1][variable]], bins=30, alpha=0.5, label=['0', '1'])
        plt.title(f'Distribution de {variable} pour CŒUR 0 et 1')
        plt.xlabel(variable)
        plt.ylabel('Fréquence')
        plt.legend()
        plt.show()
```

Analyse de corrélation

```python
import matplotlib.pyplot as plt

# Calculer la matrice de corrélation
matrice_corr = df.corr()

# Créer un heatmap pour visualiser la corrélation
plt.figure(figsize=(10, 8))
sns.heatmap(matrice_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de corrélation')
plt.show()
```
