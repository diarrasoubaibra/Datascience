```python
import re
import string
from unidecode import unidecode
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# installer le modèle spaCy pour le français
# python -m spacy download fr_core_news_sm
nlp = spacy.load('fr_core_news_sm')

# Téléchargez les ressources nécessaires pour nltk
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')

# Liste d'expressions favorables et défavorables
expressions_favorables = [
    'Bravo', 'Merci', 'Excellente initiative', 'Félicitations', 'Grande vision', 'Détermination',
    'Leadership fort', 'Admirable', 'Engagé', 'Support total', 'Réussites impressionnantes',
    'Changement positif', 'Développement durable', 'Fier de notre leader', 'Honneur', 'Respect',
    'Confiance', 'Énergie positive', 'Optimiste pour l\'avenir', 'Grand président', 'dynamique'
]

expressions_defavorables = [
    'Déception', 'Inquiétant', 'Échec', 'Manque de leadership', 'Problèmes persistants',
    'Mauvaise gestion', 'Controversé', 'Défauts évidents', 'Déclin', 'Déficit de confiance',
    'Insatisfaisant', 'Promesses non tenues', 'Politique inefficace', 'Désaccord profond',
    'Réactions négatives', 'Insuffisant', 'Incompétence', 'Préoccupant', 'Scepticisme généralisé',
    'Mauvaises décisions'
]

# Initialiser le lemmatiseur de nltk
lemmatiseur = WordNetLemmatizer()

# Liste pour stocker les commentaires nettoyés
commentaires_nettoyes = []

for commentaire in commentaires:
    commentaire = re.sub(f"[{string.punctuation}]", "", commentaire)
    commentaire = unidecode(commentaire)
    mots = nltk.word_tokenize(commentaire)
    mots = [mot for mot in mots if mot not in stopwords.words('french')]

    # Utilisation de nltk pour la lemmatisation
    commentaire_nettoye = " ".join([lemmatiseur.lemmatize(mot) for mot in mots])

    # Supprimer les lettres non prises en charge par la lemmatisation
    commentaire_nettoye = re.sub(r'\b\w{1,2}\b', '', commentaire_nettoye)

    commentaires_nettoyes.append(commentaire_nettoye)



# Analyse de sentiment à l'aide de NLTK
sia = SentimentIntensityAnalyzer()

# Liste pour stocker les résultats de l'analyse de sentiment de chaque avis
resultats_sentiment_nltk = []

# Analyse de sentiment pour chaque avis
for avis in commentaires_nettoyes:
    sentiment = sia.polarity_scores(avis)
    resultats_sentiment_nltk.append(sentiment)

# Mots-clés associés aux avis positifs
mots_cles_positifs = ["positif", "soutien", "favorable", "content", "satisfait", "terminer", "rester", "bon", "intelligent"
'Bravo', 'Merci', 'Excellente initiative', 'Félicitations', 'Grande vision', 'Détermination',
    'Leadership fort', 'Admirable', 'Engagé', 'Support total', 'Réussites impressionnantes',
    'Changement positif', 'Développement durable', 'Fier de notre leader', 'Honneur', 'Respect',
    'Confiance', 'Énergie positive', 'Optimiste pour l\'avenir', 'Grand président', 'dynamique', "Président exemplaire",
                      "bonnes relations", "Que dieu vous bénisse", "mobilisation"]

# Mots-clés associés aux avis négatifs
mots_cles_negatifs = ["dictature",  "contre", "désaccord", "problème",  "déception", "insatisfait", "famine", "chaumage"
                      'Déception', 'Inquiétant', 'Échec', 'Manque de leadership', 'Problèmes persistants',
    'Mauvaise gestion', 'Controversé', 'Défauts évidents', 'Déclin', 'Déficit de confiance',
    'Insatisfaisant', 'Promesses non tenues', 'Politique inefficace', 'Désaccord profond',
    'Réactions négatives', 'Insuffisant', 'Incompétence', 'Préoccupant', 'Scepticisme généralisé',
    'Mauvaises décisions']

# Liste pour stocker les polarités des avis basées sur les mots-clés
polarites_avis_mots_cles = []

# Analyse de chaque avis basée sur les mots-clés
for avis in commentaires_nettoyes:
    avis = avis.lower()

    # Compter le nombre de mots-clés positifs et négatifs
    count_positifs = sum(1 for mot in mots_cles_positifs if mot in avis)
    count_negatifs = sum(1 for mot in mots_cles_negatifs if mot in avis)

    # Attribuer une polarité basée sur les mots-clés
    if count_positifs > count_negatifs:
        polarite = "positif"
    elif count_negatifs > count_positifs:
        polarite = "négatif"
    else:
        polarite = "neutre"

    polarites_avis_mots_cles.append(polarite)

# Combiner les polarités basées sur le sentiment et les mots-clés
polarites_avis_combinees = []

for i in range(len(resultats_sentiment_nltk)):
    polarite_sentiment = resultats_sentiment_nltk[i]
    polarite_mots_cles = polarites_avis_mots_cles[i]

    # Logique de combinaison des polarités (à personnaliser)
    if polarite_sentiment['compound'] >= 0.05:
        polarite_combinee = "positif"
    else:
        polarite_combinee = polarite_mots_cles

    polarites_avis_combinees.append(polarite_combinee)

# Afficher les polarités attribuées aux avis
for i, (commentaire, polarite) in enumerate(zip(commentaires, polarites_avis_combinees), 1):
    print(f"Commentaire original : {commentaire}")
    print(f"Commentaire {i}: {polarite}")

```
