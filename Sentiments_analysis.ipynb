```python
import re
import string
from unidecode import unidecode
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# installer le mod√®le spaCy pour le fran√ßais
# python -m spacy download fr_core_news_sm
nlp = spacy.load('fr_core_news_sm')

# T√©l√©chargez les ressources n√©cessaires pour nltk
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')

# Liste d'expressions favorables et d√©favorables
expressions_favorables = [
    'Bravo', 'Merci', 'Excellente initiative', 'F√©licitations', 'Grande vision', 'D√©termination',
    'Leadership fort', 'Admirable', 'Engag√©', 'Support total', 'R√©ussites impressionnantes',
    'Changement positif', 'D√©veloppement durable', 'Fier de notre leader', 'Honneur', 'Respect',
    'Confiance', '√ânergie positive', 'Optimiste pour l\'avenir', 'Grand pr√©sident', 'dynamique'
]

expressions_defavorables = [
    'D√©ception', 'Inqui√©tant', '√âchec', 'Manque de leadership', 'Probl√®mes persistants',
    'Mauvaise gestion', 'Controvers√©', 'D√©fauts √©vidents', 'D√©clin', 'D√©ficit de confiance',
    'Insatisfaisant', 'Promesses non tenues', 'Politique inefficace', 'D√©saccord profond',
    'R√©actions n√©gatives', 'Insuffisant', 'Incomp√©tence', 'Pr√©occupant', 'Scepticisme g√©n√©ralis√©',
    'Mauvaises d√©cisions'
]

# Initialiser le lemmatiseur de nltk
lemmatiseur = WordNetLemmatizer()

# Liste pour stocker les commentaires nettoy√©s
commentaires_nettoyes = []

for commentaire in commentaires:
    commentaire = re.sub(f"[{string.punctuation}]", "", commentaire)
    commentaire = unidecode(commentaire)
    mots = nltk.word_tokenize(commentaire)
    mots = [mot for mot in mots if mot not in stopwords.words('french')]

    # Utilisation de nltk pour la lemmatisation
    commentaire_nettoye = " ".join([lemmatiseur.lemmatize(mot) for mot in mots])

    # Supprimer les lettres non prises en charge par la lemmatisation
    commentaire_nettoye = re.sub(r'\b\w{1,2}\b', '', commentaire_nettoye)

    commentaires_nettoyes.append(commentaire_nettoye)



# Analyse de sentiment √† l'aide de NLTK
sia = SentimentIntensityAnalyzer()

# Liste pour stocker les r√©sultats de l'analyse de sentiment de chaque avis
resultats_sentiment_nltk = []

# Analyse de sentiment pour chaque avis
for avis in commentaires_nettoyes:
    sentiment = sia.polarity_scores(avis)
    resultats_sentiment_nltk.append(sentiment)

# Mots-cl√©s associ√©s aux avis positifs
mots_cles_positifs = ["positif", "soutien", "favorable", "content", "satisfait", "terminer", "rester", "bon", "intelligent"
'Bravo', 'Merci', 'Excellente initiative', 'F√©licitations', 'Grande vision', 'D√©termination',
    'Leadership fort', 'Admirable', 'Engag√©', 'Support total', 'R√©ussites impressionnantes',
    'Changement positif', 'D√©veloppement durable', 'Fier de notre leader', 'Honneur', 'Respect',
    'Confiance', '√ânergie positive', 'Optimiste pour l\'avenir', 'Grand pr√©sident', 'dynamique', "Pr√©sident exemplaire",
                      "bonnes relations", "Que dieu vous b√©nisse", "mobilisation"]

# Mots-cl√©s associ√©s aux avis n√©gatifs
mots_cles_negatifs = ["dictature",  "contre", "d√©saccord", "probl√®me",  "d√©ception", "insatisfait", "famine", "chaumage"
                      'D√©ception', 'Inqui√©tant', '√âchec', 'Manque de leadership', 'Probl√®mes persistants',
    'Mauvaise gestion', 'Controvers√©', 'D√©fauts √©vidents', 'D√©clin', 'D√©ficit de confiance',
    'Insatisfaisant', 'Promesses non tenues', 'Politique inefficace', 'D√©saccord profond',
    'R√©actions n√©gatives', 'Insuffisant', 'Incomp√©tence', 'Pr√©occupant', 'Scepticisme g√©n√©ralis√©',
    'Mauvaises d√©cisions']

# Liste pour stocker les polarit√©s des avis bas√©es sur les mots-cl√©s
polarites_avis_mots_cles = []

# Analyse de chaque avis bas√©e sur les mots-cl√©s
for avis in commentaires_nettoyes:
    avis = avis.lower()

    # Compter le nombre de mots-cl√©s positifs et n√©gatifs
    count_positifs = sum(1 for mot in mots_cles_positifs if mot in avis)
    count_negatifs = sum(1 for mot in mots_cles_negatifs if mot in avis)

    # Attribuer une polarit√© bas√©e sur les mots-cl√©s
    if count_positifs > count_negatifs:
        polarite = "positif"
    elif count_negatifs > count_positifs:
        polarite = "n√©gatif"
    else:
        polarite = "neutre"

    polarites_avis_mots_cles.append(polarite)

# Combiner les polarit√©s bas√©es sur le sentiment et les mots-cl√©s
polarites_avis_combinees = []

for i in range(len(resultats_sentiment_nltk)):
    polarite_sentiment = resultats_sentiment_nltk[i]
    polarite_mots_cles = polarites_avis_mots_cles[i]

    # Logique de combinaison des polarit√©s (√† personnaliser)
    if polarite_sentiment['compound'] >= 0.05:
        polarite_combinee = "positif"
    else:
        polarite_combinee = polarite_mots_cles

    polarites_avis_combinees.append(polarite_combinee)

# Afficher les polarit√©s attribu√©es aux avis
for i, (commentaire, polarite) in enumerate(zip(commentaires, polarites_avis_combinees), 1):
    print(f"Commentaire original : {commentaire}")
    print(f"Commentaire {i}: {polarite}")

```
